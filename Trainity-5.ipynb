```python
data=data.drop(['num_critic_for_reviews','num_user_for_reviews','director_facebook_likes','content_rating','movie_facebook_likes'],axis=1)
```

```python
data['gross']=data['gross'].fillna((data['gross'].mean()))
```

```python
data['director_name']=data['director_name'].fillna(value='-')
```

```python
data['country']=data['country'].fillna(data['country'].mode().iloc[0])
```

```python
data['actor_1_name']=data['actor_1_name'].fillna(value=0)
```
